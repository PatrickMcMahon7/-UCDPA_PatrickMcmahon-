```python
import re
import pandas as pd
import numpy as np
```

```python
# Load the data into a DataFrame
file_path = 'C:/Users/patri/OneDrive/Documents/PYthon course/files/tested.csv'
df = pd.read_csv(file_path)

# Replace missing values with a specific value
df.fillna(value='N/A', inplace=True)

# Drop duplicates
df.drop_duplicates(inplace=True)
```

```python
# Convert Age column to a numpy array with NaN for non-numeric values
age_arr = pd.to_numeric(df['Age'], errors='coerce').to_numpy()

# Mean age
mean_age = np.nanmean(age_arr)
print(f"Mean Age: {mean_age}")
```

```python
# Define the regex pattern
pattern = r',\s*([^\.]*)\s*\.'

# Extract the pattern using regex
extracted_titles = df['Name'].str.extract(pattern, expand=False)
```

```python
if extracted_titles.notnull().any():
    print("Pattern found")
else:
    print("Pattern not found")

print(extracted_titles)    
```

```python
# Iterate through the data using iterators
for index, row in df.iterrows():
    # Access data for each row using row['column_name']
    passenger_id = row['PassengerId']
    name = row['Name']
    age = row['Age']
    

    #passenger information
    print(f"Passenger ID: {passenger_id}")
    print(f"Name: {name}")
    print(f"Age: {age}")
```
